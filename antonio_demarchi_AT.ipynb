```python
#1- Responda:
print("1- Responda:")
print("a")
#a) No contexto de ciência de dados, o que é uma feature? Por que as features precisam ser tratadas antes de serem aplicadas nos algoritmos de Machine Learning?


#Uma feature (característica) é uma variável explicativa que mostra um atributo quantificável/categórico dos dados, descrevendo um aspecto do objeto ou evento a ser analisado.
#Como se fosse as colunas em uma tabela, onde cada linha é uma instância e cada feature descreve uma característica relevante.
#Podendo ser obtidas diretamente dos dados brutos ou criadas por meio de transformações e combinações.
#A qualidade e a escolha das features impactam diretamente o desempenho do modelo. Features importantes permitem que o modelo aprenda padrões, carregando informação a partir
#da qual os algoritmos inferem relações.
#Features desnecessárias, com ruído ou valores null podem prejudicar as relações estatísticas, levando a problemas como o overfitting (memorizando os dados de treino em vez de generalizar).
#Exemplo: No dataset Palmer Penguins, features como o tamanho do bico, peso, espécie e ilha de origem são detalhes sobre os pinguins.
#Algumas são numéricas, outras categóricas (sexo, espécie).
#Nem todas tem a mesma utilidade, o comprimento e a profundidade do bico, por exemplo, são mais relevantes para diferenciar espécies do que o ano de observação.

#O tratamento das features (pré-processamento) é muito util antes de aplicá-las em algoritmos de ML:
#- Heterogeneidade de Tipos de Dados: Algoritmos esperam tipos de dados específicos sendo necessario como por exemplo converter texto em números.
#- Variabilidade de Escalas e Distribuição: Diferentes escalas (idade anos, renda em milhares) podem fazer com que features com valores maiores dominem o treinamento. Normalização e padronização corrigem isso.
#- Valores Faltantes e Outliers: Dados incompletos ou anômalos podem distorcer os resultados. É preciso imputar valores ou remover essas instâncias.
#- Redução de Dimensionalidade e Seleção de Features: Muitos recursos podem levar a modelos mais lentos e com maior chance de overfitting. Técnicas de seleção e redução ajudam a focar nas informações mais importantes.
#- Representatividade e Informação Estatística: Fazer com que as features representam bem o fenômeno real e que carregam a informação estatística necessária para o aprendizado.
#A engenharia de features é o processo de formular as features mais apropriadas, considerando os dados, o modelo e a tarefa em questão.


####################################################################################################################################################################################################################################

print("b")
#b) Sabendo que nem todas as features em um dataset são úteis para os modelos de Machine Learning, explique como as features de fato úteis podem ser selecionadas através dos métodos de filtragem, wrapper e embedding.

#Nem todas as features em um dataset são úteis para modelos de Machine Learning. A inserção de variáveis ruins para determinado contexto pode prejudicar o desempenho, aumentar o risco de overfitting, aumentar custo de operação, dificultar a interpretação dos resultados.
#Uma boa seleção de features ajuda a melhorar a performance, otimizar modelos, simplificar e torná-lo mais interpretável.
#Existem três formas principais para essa seleção: [filtragem, wrapper e embedding]
#Métodos de Filtragem:
#São métodos independentes do modelo, que avaliam cada feature com base em métricas estatísticas, sem envolver algoritmo de machine learning. Eles selecionam as variáveis mais relevantes com base em critérios como correlação, teste Qui-quadrado, Mutual Information, ou teste ANOVA.
#São rápidos e adequados para datasets grandes, pois não exigem treinamento de modelo.
#Desconsideram interações entre features, podendo eliminar variáveis relevantes em combinação com outras.
#Métodos Wrapper:
#Esses métodos avaliam o desempenho de modelos sucessivamente treinados com diferentes subconjuntos de features.
#A cada iteração, um modelo é treinado e validado, e as combinações que levam ao melhor desempenho (acurácia, F1-score, ROC) são selecionadas.
#Como Recursive Feature Elimination, Backward Elimination.
#Consideram interações entre as variáveis e geralmente levam a um desempenho superior.
#Possuem alto custo computacional, pois requerem múltiplas execuções do algoritmo de machine learning. São mais indicados para conjuntos de dados menores, onde o custo é viável.
#Métodos Embedding:
#Nos métodos embedded, a seleção de features é incorporada ao próprio processo de treinamento do modelo. As variáveis são avaliadas e selecionadas com base na estrutura interna do algoritmo, como acontece com a regularização L1 (Lasso), Árvores de Decisão, Florestas Aleatórias e métodos de boosting (XGBoost).
#Apresentam equilíbrio entre eficiência e precisão. Não necessitam de processamento anterior nem de execuções repetidas, pois o modelo atribui pesos ou importâncias às variáveis durante o próprio treinamento, identificando as mais relevantes automaticamente.
#Dependem do tipo de algoritmo utilizado.
#Nem todos os modelos possuem mecanismos internos para seleção de variáveis.
#Utilizando todas as features: Pode entregar boa acurácia, mas traz maior complexidade e risco de overfitting.
#Apenas as features selecionadas por filtragem estatística: Reduz a dimensionalidade e mantém desempenho bom, mas pode perder interações importantes.
#Apenas as features selecionadas por wrapper: Desempenho próximo ao filtro, mas com validação baseada em performance do modelo, a um custo computacional maior.
#Aplicando técnicas de redução como PCA (extra): Reduz muito as dimensões e pode, em alguns casos, prejudicar o desempenho se ignorar variáveis essenciais.

####################################################################################################################################################################################################################################


print("c")
#c) Explique com suas palavras o que são escalares, vetores e espaços vetoriais. Dê a sua resposta ilustrando cada conceito com dados usados em algoritmos de Machine Learning.


#Escalar:
#Um escalar é um elemento de um corpo numérico, que normalmente pertencente ao conjunto dos reais, uma medida.
#Aparecem normalmente em métricas, parâmetros dos modelos, e nas operações de normalização e padronização de dados.

#Exemplo: Ao calcular a acurácia de um classificador, erro quadrático médio em um problema de regressão, o resultado final é um escalar que resume o desempenho do modelo.
#Operações de pré-processamento como a padronização de variáveis transformam cada atributo ao subtrair a média e dividir pelo desvio padrão, ambos valores escalares.

#from sklearn.preprocessing import StandardScaler
#scaler = StandardScaler()
#X_scaled = scaler.fit_transform(X)

#Vetores:
#Um vetor é uma coleção ordenada de escalares, que podem representar um ponto em um espaço multi-dimensional.
#Cada instância/amostra de um dataset normalmente é representada por um vetor, onde cada componente corresponde ao valor de uma feature.
#Exemplo: Um problema de classificação de imagens, cada imagem pode ser representada como um vetor no qual os componentes são os pixels da imagem/características extraídas.
#Vetores permitem modelar dados multidimensionais e possibilitam a aplicação de operações matemáticas, como a soma vetorial, produto escalar, projeções e transformações lineares.

#Em um case de com dataset de clientes bancários, poderíamos usar um vetor de características onde os componentes representam idade, renda, histórico de crédito e presença/ausência de inadimplência.

#Espaços Vetoriais
##O espaço vetorial é a estrutura algébrica formada por todos os possíveis vetores que podem ser criados a partir de elementos de uma base, considerando que as operações de adição vetorial e multiplicação por escalar segue determinados parâmetros (como associatividade, distributividade, existência de vetor nulo, etc).
#Os espaços vetoriais definem o domínio matemático de todas as combinações possíveis dos dados de entrada.
#Ajudando no funcionamento dos algoritmos, já que as operações como a redução de dimensionalidade (PCA), projeções (transformações lineares), cálculo de distâncias (métricas de similaridade) e decomposições matriciais dependem inteiramente da noção de espaço vetorial.



####################################################################################################################################################################################################################################






```

```python
print("5- Considere o Large Movie Review Dataset (Large Movie Review Dataset )")

print("a- Remova as stopwords das resenhas do dataset.")

import tarfile, urllib.request, os, nltk, re, pickle
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from joblib import Parallel, delayed
import spacy
from collections import Counter

#download e extrcao do dataset
url = "https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"
filename = "aclImdb_v1.tar.gz"
urllib.request.urlretrieve(url, filename)
with tarfile.open(filename, "r:gz") as tar:
    tar.extractall()

#configuracao inicial
nltk.download('stopwords')
nltk.download('punkt')
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])
stop_words = set(stopwords.words('english'))

#funcao para carregar arquivos com segurança
def load_file_safe(filepath):
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            return f.read()
    except (UnicodeDecodeError, IOError) as e:
        print(f"Erro ao ler {filepath}: {e}")
        return ""

#caarregamento paralelo de resenhas
def load_reviews_parallel(data_dir):
    pos_dir = os.path.join(data_dir, 'pos')
    neg_dir = os.path.join(data_dir, 'neg')
    pos_files = [os.path.join(pos_dir, f) for f in os.listdir(pos_dir)]
    neg_files = [os.path.join(neg_dir, f) for f in os.listdir(neg_dir)]
    reviews = Parallel(n_jobs=-1)(delayed(load_file_safe)(f) for f in pos_files + neg_files)
    return reviews

#pree-processamento com spaCy
def preprocess_text_spacy(text):
    text = re.sub(r'<br />', ' ', text).lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    doc = nlp(text)
    return " ".join([token.text for token in doc if token.text not in stop_words])

#carregar resenhas
train_reviews = load_reviews_parallel('aclImdb/train')
test_reviews = load_reviews_parallel('aclImdb/test')
print(f"Total de resenhas de treino: {len(train_reviews)}")
print(f"Total de resenhas de teste: {len(test_reviews)}")

#processar resenhas
processed_train_reviews = [preprocess_text_spacy(review) for review in train_reviews]
processed_test_reviews = [preprocess_text_spacy(review) for review in test_reviews]

#exibir exemplo
print("Exemplo Original (Treino)")
print(train_reviews[0][:300])
print("Exemplo Processada (Treino)")
print(processed_train_reviews[0][:300])

#analise de frequencia
print("Frequência Palavras")
print("Antes:", Counter(" ".join(train_reviews).split()).most_common(10))
print("Depois:", Counter(" ".join(processed_train_reviews).split()).most_common(10))

#salvar resultados
with open('processed_train_reviews.pkl', 'wb') as f:
    pickle.dump(processed_train_reviews, f)
with open('processed_test_reviews.pkl', 'wb') as f:
    pickle.dump(processed_test_reviews, f)


##################################################################################

print("b- Realize o stemming das expressões contidas nas features resultantes da operação realizada no item (a).")

from nltk.stem import SnowballStemmer

#configuracao do stemmer
stemmer = SnowballStemmer("english")

#funcao de preprocessamento com stemming
def preprocess_text_spacy_with_stemming(text):
    text = re.sub(r'<br />', ' ', text).lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    doc = nlp(text)
    stemmed_words = [stemmer.stem(token.text) for token in doc if token.text not in stop_words]
    return " ".join(stemmed_words)

#processar resenhas com stemming
processed_train_reviews_stemmed = [preprocess_text_spacy_with_stemming(review) for review in train_reviews]
processed_test_reviews_stemmed = [preprocess_text_spacy_with_stemming(review) for review in test_reviews]

#exibir exemplo
print("\n--- Resenha Processada com Stopwords (Treino) ---")
print(processed_train_reviews[0][:300])
print("\n--- Resenha com Stopwords e Stemming (Treino) ---")
print(processed_train_reviews_stemmed[0][:300])

#frequencia de palavras
print("\n--- Frequência de Palavras ---")
print("Antes Stemming:", Counter(" ".join(processed_train_reviews).split()).most_common(10))
print("Depois Stemming:", Counter(" ".join(processed_train_reviews_stemmed).split()).most_common(10))

#saalvar resultados
with open('processed_train_reviews_stemmed.pkl', 'wb') as f:
    pickle.dump(processed_train_reviews_stemmed, f)
with open('processed_test_reviews_stemmed.pkl', 'wb') as f:
    pickle.dump(processed_test_reviews_stemmed, f)

print("c- Realize a lemmatization das expressões contidas nas features resultantes da operação realizada no item (a)")

import re
import spacy
from collections import Counter
import pickle

#configuracao do spaCy
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])
stop_words = set(stopwords.words('english'))

#funcao de pre-processamento com lematizacao
def preprocess_text_spacy_with_lemmatization(text):
    text = re.sub(r'<br />', ' ', text).lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    doc = nlp(text)
    lemmatized_words = [token.lemma_ for token in doc if token.text not in stop_words]
    return " ".join(lemmatized_words)

#processar resenhas com lematização
processed_train_reviews_lemmatized = [preprocess_text_spacy_with_lemmatization(review) for review in train_reviews]
processed_test_reviews_lemmatized = [preprocess_text_spacy_with_lemmatization(review) for review in test_reviews]

#exibir exemplo
print("\n--- Resenha Processada com Stopwords (Treino) ---")
print(processed_train_reviews[0][:300])
print("\n--- Resenha com Stopwords e Lematização (Treino) ---")
print(processed_train_reviews_lemmatized[0][:300])

#frequencia de palavras
print("\n--- Frequência de Palavras ---")
print("Antes Lematização:", Counter(" ".join(processed_train_reviews).split()).most_common(10))
print("Depois Lematização:", Counter(" ".join(processed_train_reviews_lemmatized).split()).most_common(10))

#salvar resultados
with open('processed_train_reviews_lemmatized.pkl', 'wb') as f:
    pickle.dump(processed_train_reviews_lemmatized, f)
with open('processed_test_reviews_lemmatized.pkl', 'wb') as f:
    pickle.dump(processed_test_reviews_lemmatized, f)


print("d- Utilize a técnica de Bag-of-Words para vetorizar as resenhas do dataset resultante das operações do item (b). Apresente os resultados obtidos.")

from sklearn.feature_extraction.text import CountVectorizer
import pickle
import numpy as np

#inicializar o CountVectorizer
vectorizer = CountVectorizer(max_features=5000, binary=False)  # Limite de 5000 palavras mais frequentes

#vetorizar resenhas de treino e teste (usando stemmed reviews do item b)
X_train_bow = vectorizer.fit_transform(processed_train_reviews_stemmed)
X_test_bow = vectorizer.transform(processed_test_reviews_stemmed)

#obter nomes das features (vocabulário)
feature_names = vectorizer.get_feature_names_out()

#exibir resultados
print("\n--- Resultados da Vetorização Bag-of-Words ---")
print(f"Tamanho da matriz de treino (n_samples, n_features): {X_train_bow.shape}")
print(f"Tamanho da matriz de teste (n_samples, n_features): {X_test_bow.shape}")
print(f"Tamanho do vocabulário: {len(feature_names)}")
print(f"Exemplo de palavras no vocabulário (primeiras 10): {feature_names[:10]}")
print(f"Exemplo de vetor BoW (primeira resenha, primeiras 10 features): {X_train_bow[0].toarray()[0][:10]}")

#salvar matrizes e vectorizer
with open('X_train_bow.pkl', 'wb') as f:
    pickle.dump(X_train_bow, f)
with open('X_test_bow.pkl', 'wb') as f:
    pickle.dump(X_test_bow, f)
with open('vectorizer_bow.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)


print("e- Utilize a técnica de Bag-of-n-Grams, identificando bigrams, para vetorizar as resenhas do dataset resultante das operações do item (c). Apresente os resultados obtidos.")

from sklearn.feature_extraction.text import CountVectorizer
import pickle
import numpy as np

#inicializar o countVectorizer para bigrams
vectorizer_bigrams = CountVectorizer(ngram_range=(2, 2), max_features=5000)  # Apenas bigrams, limite de 5000 features

#vetorizar resenhas de treino e teste (lemmatized item c)
X_train_bigrams = vectorizer_bigrams.fit_transform(processed_train_reviews_lemmatized)
X_test_bigrams = vectorizer_bigrams.transform(processed_test_reviews_lemmatized)

#obter nomes das features (vocabulário de bigrams)
feature_names_bigrams = vectorizer_bigrams.get_feature_names_out()

#exibir resultados
print("\n--- Resultados da Vetorização Bag-of-n-Grams (Bigrams) ---")
print(f"Tamanho da matriz de treino (n_samples, n_features): {X_train_bigrams.shape}")
print(f"Tamanho da matriz de teste (n_samples, n_features): {X_test_bigrams.shape}")
print(f"Tamanho do vocabulário de bigrams: {len(feature_names_bigrams)}")
print(f"Exemplo de bigrams no vocabulário (primeiros 10): {feature_names_bigrams[:10]}")
print(f"Exemplo de vetor de bigrams (primeira resenha, primeiras 10 features): {X_train_bigrams[0].toarray()[0][:10]}")

#salvar matrizes e vectorizer
with open('X_train_bigrams.pkl', 'wb') as f:
    pickle.dump(X_train_bigrams, f)
with open('X_test_bigrams.pkl', 'wb') as f:
    pickle.dump(X_test_bigrams, f)
with open('vectorizer_bigrams.pkl', 'wb') as f:
    pickle.dump(vectorizer_bigrams, f)

```

```python
print("6- Considere o Large Movie Review Dataset (Large Movie Review Dataset")


print("a) Redimensione os dados do dataset pelo método term frequency–inverse document frequency (tf-idf). Apresente os resultados obtidos.")

import tarfile, urllib.request, os, nltk, re, pickle
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from joblib import Parallel, delayed
import spacy
from collections import Counter

#download e extrcao do dataset
url = "https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"
filename = "aclImdb_v1.tar.gz"
urllib.request.urlretrieve(url, filename)
with tarfile.open(filename, "r:gz") as tar:
    tar.extractall()

#configuracao inicial
nltk.download('stopwords')
nltk.download('punkt')
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])
stop_words = set(stopwords.words('english'))


##################################################################
##################################################################
##################################################################


print("6- Considere o Large Movie Review Dataset (Large Movie Review Dataset")
print("b) Crie um modelo de classificação baseado em regressão logística sobre a base redimensionada no item (a) e avalie os resultados obtidos.")

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
import os
from joblib import Parallel, delayed

# Função para carregar resenhas e labels
def load_reviews_parallel_with_labels(data_dir):
    """Carrega resenhas e labels (1 para pos, 0 para neg) em paralelo."""
    pos_dir = os.path.join(data_dir, 'pos')
    neg_dir = os.path.join(data_dir, 'neg')
    pos_files = [(os.path.join(pos_dir, f), 1) for f in os.listdir(pos_dir)]
    neg_files = [(os.path.join(neg_dir, f), 0) for f in os.listdir(neg_dir)]
    files = pos_files + neg_files

    def load_file_safe_with_label(filepath_label):
        filepath, label = filepath_label
        try:
            with open(filepath, 'r', encoding='utf-8') as f:
                return f.read(), label
        except (UnicodeDecodeError, IOError) as e:
            print(f"Erro ao ler {filepath}: {e}")
            return "", label

    results = Parallel(n_jobs=-1)(delayed(load_file_safe_with_label)(f) for f in files)
    reviews, labels = zip(*results)
    return list(reviews), list(labels)

# Carregar resenhas e labels
print("\n--- Carregando resenhas e labels ---")
train_reviews, y_train = load_reviews_parallel_with_labels('aclImdb/train')
test_reviews, y_test = load_reviews_parallel_with_labels('aclImdb/test')
print(f"Total de resenhas de treino: {len(train_reviews)}")
print(f"Total de resenhas de teste: {len(test_reviews)}")

# Treinar modelo de regressão logística
print("\n--- Item b: Treinamento e Avaliação do Modelo de Regressão Logística ---")
clf = LogisticRegression(max_iter=1000, random_state=42)
clf.fit(X_train_tfidf, y_train)

# Prever no conjunto de teste
y_pred = clf.predict(X_test_tfidf)

# Avaliar o modelo
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Acurácia: {accuracy:.4f}")
print(f"Precisão: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# Matriz de confusão
cm = confusion_matrix(y_test, y_pred)
print("\nMatriz de Confusão:")
print(cm)

# Visualizar matriz de confusão
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negativo', 'Positivo'], yticklabels=['Negativo', 'Positivo'])
plt.xlabel('Predito')
plt.ylabel('Verdadeiro')
plt.title('Matriz de Confusão - Regressão Logística')
plt.show()

# Salvar o modelo
with open('logistic_model_tfidf.pkl', 'wb') as f:
    pickle.dump(clf, f)

print("modelo treinado e salvo em 'logistic_model_tfidf.pkl'.")

##################################################################
##################################################################
##################################################################
```
