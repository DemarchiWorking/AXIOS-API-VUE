```python
print("2- Considere o breast cancer dataset.")

#O dataset Breast Cancer Wisconsin é utilizado em problemas de classificação e possui uma série de características/features que descrevem as propriedades de núcleos celulares das amostras.
#A maioria dessas features são contínuas, representando medições numéricas que podem assumir qualquer valor dentro de um determinado intervalo.
#As features neste dataset são calculadas a partir de uma imagem digitalizada de um aspirado por agulha fina de uma massa mamária.

print("a- Defina uma função personalizada que realize uma normalização nas features do dataset usando FunctionTransformer do sklearn. Apresente os resultados obtidos.")

#RESPOSTA A
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import FunctionTransformer, MinMaxScaler

def min_max_normalize(X):
    """
    Normaliza as features para o intervalo [0, 1] usando Min-Max.
    Args:
        X (numpy.ndarray): Matriz de features.
    Returns:
        numpy.ndarray: Features normalizadas.
    """
    X_min = np.min(X, axis=0)
    X_max = np.max(X, axis=0)
    #evitar divisao por zero
    denominator = X_max - X_min
    denominator[denominator == 0] = 1  # Caso raro de feature constante
    return (X - X_min) / denominator

#carregar o dataset
cancer = load_breast_cancer()
X_df = pd.DataFrame(cancer.data, columns=cancer.feature_names)

#criar o FunctionTransformer com a função personalizada
normalizer = FunctionTransformer(func=min_max_normalize, validate=False)

#aplicar a normalizacao
X_normalized = normalizer.fit_transform(X_df)

#criar DataFrame com valores normalizados
X_normalized_df = pd.DataFrame(X_normalized, columns=cancer.feature_names)

#validar com MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_df)
X_scaled_df = pd.DataFrame(X_scaled, columns=cancer.feature_names)

#apresentar resultados
print("resultados da normalizacao com FunctionTransformer")

#amostra dos valores originais e normalizados (5)
print("\nValores Originais (primeiras 5 linhas):")
print(X_df.head().to_string())
print("\nValores Normalizados com FunctionTransformer (primeiras 5 linhas):")
print(X_normalized_df.head().to_string())

#estatisticas descritivas para verificar o intervalo [0, 1]
print("\nEstatísticas Descritivas das Features Normalizadas:")
print(X_normalized_df.describe().to_string())

#verificar consistencia com MinMaxScaler
difference = np.abs(X_normalized - X_scaled)
print("\nDiferença Máxima entre FunctionTransformer e MinMaxScaler:")
print(np.max(difference))

#contagem de valores fora do intervalo [0, 1]
out_of_range = np.sum((X_normalized < 0) | (X_normalized > 1))
print(f"\nNúmero de valores fora do intervalo [0, 1]: {out_of_range}")






# RESPOSTA B
print("b- Utilize o PowerTransform para normalizar as features do dataset. Apresente os resultados obtidos.")


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import PowerTransformer, StandardScaler
import seaborn as sns

#carregar o dataset
cancer = load_breast_cancer()
X_df = pd.DataFrame(cancer.data, columns=cancer.feature_names)

#verificar dados ausentes e tipos
print("verificacao inicial do dataset:")
print(X_df.info())
print("\n valores ausentes:", X_df.isnull().sum().sum())

#definir o PowerTransformer com Yeo-Johnson
pt = PowerTransformer(method='yeo-johnson', standardize=True)

#aplicar a transformacao
X_transformed = pt.fit_transform(X_df)
X_transformed_df = pd.DataFrame(X_transformed, columns=cancer.feature_names)

#validar com StandardScaler (comparar a padronizacao)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_df)
X_scaled_df = pd.DataFrame(X_scaled, columns=cancer.feature_names)

#apresentar resultados
print("\n=== Resultados da Normalização com PowerTransformer (Yeo-Johnson) ===")

#amostra dos valores originais e transformados
print("\n valores originais (5):")
print(X_df.head().to_string())
print("\n valores normalizados com PowerTransformer (5):")
print(X_transformed_df.head().to_string())

#estatisticas descritivas
print("\n estatisticas descritivas das features normalizadas:")
print(X_transformed_df.describe().to_string())

# Visualizar distribuições antes e depois (para duas features representativas)
features_to_plot = ['mean radius', 'mean texture']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Antes da transformação
for i, feature in enumerate(features_to_plot):
    sns.histplot(X_df[feature], kde=True, ax=axes[0, i])
    axes[0, i].set_title(f'Distribuição Original - {feature}')
    axes[0, i].set_xlabel(feature)
    axes[0, i].set_ylabel('Frequência')

# Depois da transformação
for i, feature in enumerate(features_to_plot):
    sns.histplot(X_transformed_df[feature], kde=True, ax=axes[1, i])
    axes[1, i].set_title(f'Distribuição Normalizada - {feature}')
    axes[1, i].set_xlabel(feature)
    axes[1, i].set_ylabel('Frequência')

plt.tight_layout()
plt.show()

# Verificar média e desvio padrão próximos de 0 e 1
print("\nMédia das Features Normalizadas (deve ser ~0):")
print(X_transformed_df.mean().describe())
print("\nDesvio Padrão das Features Normalizadas (deve ser ~1):")
print(X_transformed_df.std().describe())

# Comparação com StandardScaler
difference = np.abs(X_transformed - X_scaled)
print("\nDiferença Média entre PowerTransformer e StandardScaler:")
print(np.mean(difference))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import re
import os
import tarfile
import urllib.request
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer

import nltk
nltk.download('punkt_tab')

dataset_dir = 'aclImdb'
if not os.path.exists(dataset_dir):
    print(f"Diretório '{dataset_dir}' não encontrado. Tentando baixar e descompactar o dataset...")
    dataset_url = "http://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"
    archive_name = "aclImdb_v1.tar.gz"

    try:
        urllib.request.urlretrieve(dataset_url, archive_name)
        print(f"'{archive_name}' baixado com sucesso.")

        with tarfile.open(archive_name, "r:gz") as tar:
            tar.extractall()
        print(f"'{archive_name}' descompactado em '{dataset_dir}'.")

        os.remove(archive_name)
        print(f"Arquivo '{archive_name}' removido.")

    except Exception as e:
        print(f"Erro ao baixar ou descompactar o dataset: {e}")
        print("Por favor, baixe o dataset 'aclImdb_v1.tar.gz' manualmente, descompacte-o e")
        print(f"certifique-se de que a pasta '{dataset_dir}' esteja no mesmo diretório do seu script.")
        # Em caso de falha no download/extração, definimos df como um DataFrame vazio para evitar erros
        # e o restante do código terá verificações para lidar com isso.
        df = pd.DataFrame() # Define df vazio para evitar NameError mais tarde
else:
    print(f"Diretório '{dataset_dir}' já existe. Pulando o download do dataset.")


# --- Baixar recursos do NLTK ---
# É crucial que estes downloads ocorram ANTES do uso de funções NLTK como word_tokenize
print("Baixando recursos do NLTK...")
try:
    # 'punkt' é o tokenizador necessário para word_tokenize
    nltk.download('punkt', quiet=True)
    nltk.download('stopwords', quiet=True)
    nltk.download('wordnet', quiet=True)
    print("Recursos do NLTK baixados com sucesso.")
except Exception as e:
    print(f"Erro ao baixar recursos do NLTK: {e}")
    print("Por favor, verifique sua conexão com a internet ou tente baixar manualmente.")


# --- Função para carregar resenhas do diretório ---
def load_reviews(directory, sentiment):
    reviews = []
    if not os.path.exists(directory):
        print(f"Erro: Diretório não encontrado: {directory}")
        return pd.DataFrame({'review': [], 'sentiment': []})

    files = [f for f in os.listdir(directory) if f.endswith('.txt')]
    for file_name in files:
        file_path = os.path.join(directory, file_name)
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                reviews.append(f.read())
        except Exception as e:
            print(f"Erro ao ler o arquivo {file_path}: {e}")
    return pd.DataFrame({'review': reviews, 'sentiment': sentiment})

# --- Carregar o dataset ---
base_path = os.path.join(dataset_dir, 'train')

print(f"\nCarregando reviews do caminho base: {base_path}")
pos_reviews = load_reviews(os.path.join(base_path, 'pos'), 1)
neg_reviews = load_reviews(os.path.join(base_path, 'neg'), 0)

# Concatena os DataFrames de reviews positivas e negativas
# Verificação para garantir que os DataFrames não estão vazios após o carregamento
df = pd.DataFrame() # Inicializa df para garantir que sempre exista
if not pos_reviews.empty and not neg_reviews.empty:
    df = pd.concat([pos_reviews, neg_reviews], ignore_index=True)
    print(f"Dataset carregado: {df.shape[0]} resenhas.")
elif os.path.exists(dataset_dir): # Se o diretório existe mas as reviews não foram carregadas
    print("Um ou ambos os DataFrames (positivos/negativos) estão vazios. Verifique se há arquivos .txt nos diretórios esperados.")
else: # Se o diretório nem existe
    print("O dataset não foi carregado porque o diretório 'aclImdb' não foi encontrado ou não pôde ser baixado/descompactado.")


# --- Função para pré-processamento (com stemming) ---
def preprocess_text_stemming(text):
    text = re.sub(r'<.*?>', '', text)  # Remover tags HTML
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Manter apenas letras e espaços
    text = text.lower()  # Converter para minúsculas
    tokens = word_tokenize(text) # Esta linha depende do 'punkt'
    stop_words = set(stopwords.words('english')) - {'not'} # Manter 'not'
    tokens = [token for token in tokens if token not in stop_words]
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(token) for token in tokens]
    return ' '.join(tokens)

# --- Aplicar pré-processamento (com stemming) ---
print("\nAplicando pré-processamento (com stemming)...")
if not df.empty:
    df['stemmed_review'] = df['review'].apply(preprocess_text_stemming)
    print("Pré-processamento com stemming concluído.")
else:
    print("DataFrame vazio, não é possível aplicar pré-processamento (stemming).")


# --- Realizar a Lemmatization ---
print("\n--- Realizando Lemmatization ---")

lemmatizer = WordNetLemmatizer()

# Função para lematizar tokens
def lemmatize_text(text):
    tokens = word_tokenize(text) # Esta linha também depende do 'punkt'
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(lemmatized_tokens)

if 'stemmed_review' in df.columns and not df.empty:
    df['lemmatized_review'] = df['stemmed_review'].apply(lemmatize_text)
    print("Lemmatization concluída.")

    # Apresentar resultados da lematização
    print("\n=== Resultados da Lemmatization ===")
    print("Amostra de Resenhas Stemizadas vs. Lematizadas (primeiras 5):")
    print(df[['stemmed_review', 'lemmatized_review']].head().to_string())

    # Contagem de palavras e estatísticas
    df['word_count_stemmed'] = df['stemmed_review'].apply(lambda x: len(word_tokenize(x)))
    df['word_count_lemmatized'] = df['lemmatized_review'].apply(lambda x: len(word_tokenize(x)))

    print("\nEstatísticas de Contagem de Palavras (Stemmed vs. Lematized):")
    print("Stemmed:\n", df['word_count_stemmed'].describe().to_string())
    print("\nLematized:\n", df['word_count_lemmatized'].describe().to_string())

    # Inspeção Manual: Exemplo
    print("\nInspeção Manual: Exemplo de Stemming vs. Lemmatization (Resenha 100):")
    if len(df) > 100:
        sample_idx = 100
        original_cleaned_raw = re.sub(r'<.*?>', '', df.loc[sample_idx, 'review'])
        original_cleaned_alpha = re.sub(r'[^a-zA-Z\s]', '', original_cleaned_raw).lower()
        original_cleaned_tokens = word_tokenize(original_cleaned_alpha)
        stop_words_temp = set(stopwords.words('english')) - {'not'}
        original_cleaned_filtered = [token for token in original_cleaned_tokens if token not in stop_words_temp]

        print(f"Original (limpa, pré-stopwords): {' '.join(original_cleaned_filtered)[:150]}...")
        print(f"Stemmed: {df.loc[sample_idx, 'stemmed_review'][:150]}...")
        print(f"Lematized: {df.loc[sample_idx, 'lemmatized_review'][:150]}...")
    else:
        print("Dataset tem menos de 101 resenhas, não é possível exibir o exemplo 100.")
else:
    print("A coluna 'stemmed_review' não foi criada ou o DataFrame está vazio, pulando lematização.")


# --- Vetorização Bag-of-Words (BoW) ---
print("\n--- Vetorização Bag-of-Words (BoW) ---")

if 'lemmatized_review' in df.columns and not df.empty:
    vectorizer = CountVectorizer()
    X_bow = vectorizer.fit_transform(df['lemmatized_review'])

    print("Vetorização Bag-of-Words concluída.")

    # Apresentar resultados do BoW
    print("\n=== Resultados da Vetorização Bag-of-Words ===")
    print(f"Dimensão da Matriz BoW (Resenhas x Vocabulário): {X_bow.shape}")
    print(f"Número de termos únicos no vocabulário: {len(vectorizer.vocabulary_)}")

    print("\nExemplo de termos no vocabulário e seus índices (primeiros 10):")
    sample_vocab = dict(list(vectorizer.vocabulary_.items())[0:10])
    print(sample_vocab)

    print("\nRepresentação BoW para a primeira resenha (índices de termos e contagens):")
    first_review_bow_indices = X_bow[0].nonzero()[1]
    first_review_bow_counts = X_bow[0].data

    id_to_word = {v: k for k, v in vectorizer.vocabulary_.items()}
    first_review_bow_representation = {id_to_word[idx]: count for idx, count in zip(first_review_bow_indices, first_review_bow_counts)}

    print(first_review_bow_representation)
else:
    print("A coluna 'lemmatized_review' não foi criada ou o DataFrame está vazio, pulando vetorização BoW.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import re
import os
import tarfile
import urllib.request
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer

import nltk
nltk.download('punkt_tab')

dataset_dir = 'aclImdb'
if not os.path.exists(dataset_dir):
    print(f"Diretório '{dataset_dir}' não encontrado. Tentando baixar e descompactar o dataset...")
    dataset_url = "http://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"
    archive_name = "aclImdb_v1.tar.gz"

    try:
        urllib.request.urlretrieve(dataset_url, archive_name)
        print(f"'{archive_name}' baixado com sucesso.")

        with tarfile.open(archive_name, "r:gz") as tar:
            tar.extractall()
        print(f"'{archive_name}' descompactado em '{dataset_dir}'.")

        os.remove(archive_name)
        print(f"Arquivo '{archive_name}' removido.")

    except Exception as e:
        print(f"Erro ao baixar ou descompactar o dataset: {e}")
        print("Por favor, baixe o dataset 'aclImdb_v1.tar.gz' manualmente, descompacte-o e")
        print(f"certifique-se de que a pasta '{dataset_dir}' esteja no mesmo diretório do seu script.")
        # Em caso de falha no download/extração, definimos df como um DataFrame vazio para evitar erros
        # e o restante do código terá verificações para lidar com isso.
        df = pd.DataFrame() # Define df vazio para evitar NameError mais tarde
else:
    print(f"Diretório '{dataset_dir}' já existe. Pulando o download do dataset.")


# --- Baixar recursos do NLTK ---
# É crucial que estes downloads ocorram ANTES do uso de funções NLTK como word_tokenize
print("Baixando recursos do NLTK...")
try:
    # 'punkt' é o tokenizador necessário para word_tokenize
    nltk.download('punkt', quiet=True)
    nltk.download('stopwords', quiet=True)
    nltk.download('wordnet', quiet=True)
    print("Recursos do NLTK baixados com sucesso.")
except Exception as e:
    print(f"Erro ao baixar recursos do NLTK: {e}")
    print("Por favor, verifique sua conexão com a internet ou tente baixar manualmente.")


# --- Função para carregar resenhas do diretório ---
def load_reviews(directory, sentiment):
    reviews = []
    if not os.path.exists(directory):
        print(f"Erro: Diretório não encontrado: {directory}")
        return pd.DataFrame({'review': [], 'sentiment': []})

    files = [f for f in os.listdir(directory) if f.endswith('.txt')]
    for file_name in files:
        file_path = os.path.join(directory, file_name)
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                reviews.append(f.read())
        except Exception as e:
            print(f"Erro ao ler o arquivo {file_path}: {e}")
    return pd.DataFrame({'review': reviews, 'sentiment': sentiment})

# --- Carregar o dataset ---
base_path = os.path.join(dataset_dir, 'train')

print(f"\nCarregando reviews do caminho base: {base_path}")
pos_reviews = load_reviews(os.path.join(base_path, 'pos'), 1)
neg_reviews = load_reviews(os.path.join(base_path, 'neg'), 0)

# Concatena os DataFrames de reviews positivas e negativas
# Verificação para garantir que os DataFrames não estão vazios após o carregamento
df = pd.DataFrame() # Inicializa df para garantir que sempre exista
if not pos_reviews.empty and not neg_reviews.empty:
    df = pd.concat([pos_reviews, neg_reviews], ignore_index=True)
    print(f"Dataset carregado: {df.shape[0]} resenhas.")
elif os.path.exists(dataset_dir): # Se o diretório existe mas as reviews não foram carregadas
    print("Um ou ambos os DataFrames (positivos/negativos) estão vazios. Verifique se há arquivos .txt nos diretórios esperados.")
else: # Se o diretório nem existe
    print("O dataset não foi carregado porque o diretório 'aclImdb' não foi encontrado ou não pôde ser baixado/descompactado.")


# --- Função para pré-processamento (com stemming) ---
def preprocess_text_stemming(text):
    text = re.sub(r'<.*?>', '', text)  # Remover tags HTML
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Manter apenas letras e espaços
    text = text.lower()  # Converter para minúsculas
    tokens = word_tokenize(text) # Esta linha depende do 'punkt'
    stop_words = set(stopwords.words('english')) - {'not'} # Manter 'not'
    tokens = [token for token in tokens if token not in stop_words]
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(token) for token in tokens]
    return ' '.join(tokens)

# --- Aplicar pré-processamento (com stemming) ---
print("\nAplicando pré-processamento (com stemming)...")
if not df.empty:
    df['stemmed_review'] = df['review'].apply(preprocess_text_stemming)
    print("Pré-processamento com stemming concluído.")
else:
    print("DataFrame vazio, não é possível aplicar pré-processamento (stemming).")


# --- Realizar a Lemmatization ---
print("\n--- Realizando Lemmatization ---")

lemmatizer = WordNetLemmatizer()

# Função para lematizar tokens
def lemmatize_text(text):
    tokens = word_tokenize(text) # Esta linha também depende do 'punkt'
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(lemmatized_tokens)

if 'stemmed_review' in df.columns and not df.empty:
    df['lemmatized_review'] = df['stemmed_review'].apply(lemmatize_text)
    print("Lemmatization concluída.")

    # Apresentar resultados da lematização
    print("\n=== Resultados da Lemmatization ===")
    print("Amostra de Resenhas Stemizadas vs. Lematizadas (primeiras 5):")
    print(df[['stemmed_review', 'lemmatized_review']].head().to_string())

    # Contagem de palavras e estatísticas
    df['word_count_stemmed'] = df['stemmed_review'].apply(lambda x: len(word_tokenize(x)))
    df['word_count_lemmatized'] = df['lemmatized_review'].apply(lambda x: len(word_tokenize(x)))

    print("\nEstatísticas de Contagem de Palavras (Stemmed vs. Lematized):")
    print("Stemmed:\n", df['word_count_stemmed'].describe().to_string())
    print("\nLematized:\n", df['word_count_lemmatized'].describe().to_string())

    # Inspeção Manual: Exemplo
    print("\nInspeção Manual: Exemplo de Stemming vs. Lemmatization (Resenha 100):")
    if len(df) > 100:
        sample_idx = 100
        original_cleaned_raw = re.sub(r'<.*?>', '', df.loc[sample_idx, 'review'])
        original_cleaned_alpha = re.sub(r'[^a-zA-Z\s]', '', original_cleaned_raw).lower()
        original_cleaned_tokens = word_tokenize(original_cleaned_alpha)
        stop_words_temp = set(stopwords.words('english')) - {'not'}
        original_cleaned_filtered = [token for token in original_cleaned_tokens if token not in stop_words_temp]

        print(f"Original (limpa, pré-stopwords): {' '.join(original_cleaned_filtered)[:150]}...")
        print(f"Stemmed: {df.loc[sample_idx, 'stemmed_review'][:150]}...")
        print(f"Lematized: {df.loc[sample_idx, 'lemmatized_review'][:150]}...")
    else:
        print("Dataset tem menos de 101 resenhas, não é possível exibir o exemplo 100.")
else:
    print("A coluna 'stemmed_review' não foi criada ou o DataFrame está vazio, pulando lematização.")


# --- Vetorização Bag-of-Words (BoW) ---
print("\n--- Vetorização Bag-of-Words (BoW) ---")

if 'lemmatized_review' in df.columns and not df.empty:
    vectorizer = CountVectorizer()
    X_bow = vectorizer.fit_transform(df['lemmatized_review'])

    print("Vetorização Bag-of-Words concluída.")

    # Apresentar resultados do BoW
    print("\n=== Resultados da Vetorização Bag-of-Words ===")
    print(f"Dimensão da Matriz BoW (Resenhas x Vocabulário): {X_bow.shape}")
    print(f"Número de termos únicos no vocabulário: {len(vectorizer.vocabulary_)}")

    print("\nExemplo de termos no vocabulário e seus índices (primeiros 10):")
    sample_vocab = dict(list(vectorizer.vocabulary_.items())[0:10])
    print(sample_vocab)

    print("\nRepresentação BoW para a primeira resenha (índices de termos e contagens):")
    first_review_bow_indices = X_bow[0].nonzero()[1]
    first_review_bow_counts = X_bow[0].data

    id_to_word = {v: k for k, v in vectorizer.vocabulary_.items()}
    first_review_bow_representation = {id_to_word[idx]: count for idx, count in zip(first_review_bow_indices, first_review_bow_counts)}

    print(first_review_bow_representation)
else:
    print("A coluna 'lemmatized_review' não foi criada ou o DataFrame está vazio, pulando vetorização BoW.")
```

```python
#7-Vetorização de Variáveis Categóricas no Powerlifting Database
#A vetorização de variáveis categóricas é um processo fundamental no pré-processamento de dados para Machine Learning, convertendo atributos nominais ou ordinais em uma representação numérica. Abordaremos três métodos distintos: One-Hot Encoding, Dummy Coding e Effect Coding,


import pandas as pd
import numpy as np

# Carregar o dataset
try:
    df = pd.read_csv('openpowerlifting.csv', low_memory=False)
    print("Dataset 'openpowerlifting.csv' carregado com sucesso.")
except FileNotFoundError:
    print("Erro: 'openpowerlifting.csv' não encontrado.")
    print("Por favor, baixe o arquivo 'openpowerlifting.csv' em https://www.kaggle.com/open-powerlifting/powerlifting-database")
    print("e coloque-o no mesmo diretório deste script.")
    print("Criando um DataFrame de exemplo para demonstração das técnicas de vetorização.")
    # Criar um DataFrame de exemplo caso o arquivo não seja encontrado
    data = {
        'Sex': ['M', 'F', 'M', 'F', 'M', 'M', 'F'],
        'Equipment': ['Raw', 'Wraps', 'Raw', 'Single-ply', 'Raw', 'Wraps', 'Raw'],
        'Federation': ['IPF', 'USAPL', 'IPF', 'WPC', 'USAPL', 'IPF', 'WPC'],
        'AgeClass': ['24-34', '35-39', '24-34', '40-44', '24-34', '20-23', '35-39']
    }
    df = pd.DataFrame(data)

# Selecionar variáveis categóricas para vetorização
categorical_cols = ['Sex', 'Equipment', 'Federation']
if 'AgeClass' in df.columns: # Adicionar 'AgeClass' se a coluna existir no dataset
    categorical_cols.append('AgeClass')

# Filtrar o DataFrame para incluir apenas as colunas categóricas e remover NaNs para simplicidade
# Usar .copy() para evitar SettingWithCopyWarning e garantir um DataFrame independente
df_categorical = df[categorical_cols].dropna().head(100).copy()

print("\nAmostra das variáveis categóricas selecionadas:")
print(df_categorical.head().to_string())


#Considere o powerlifting database (powerlifting-database)

#O One-Hot Encoding é uma técnica de vetorização que cria uma nova coluna binária (dummy) para cada categoria única presente em uma variável categórica original. Para cada observação, a coluna correspondente à sua categoria recebe o valor 1 (ou True), enquanto as demais colunas dessa variável recebem 0 (ou False). Este método é amplamente utilizado pois evita a inferência de qualquer ordem ordinal entre as categorias, sendo robusto para diversos algoritmos de Machine Learning. Contudo, pode levar a um aumento significativo na dimensionalidade do dataset, especialmente com variáveis que possuem muitas categorias distintas.


#Vetorize as variáveis categóricas usando One-hot Encoding. Apresente os resultados obtidos.



#Vetorize as variáveis categóricas usando Dummy Coding. Compare os resultados desta vetorização com aqueles obtidos no item (a).


#O Dummy Coding é uma variação do One-Hot Encoding que, para cada variável categórica original com K categorias, gera K-1 novas colunas binárias. Uma das categorias é designada como "categoria de referência" e é omitida, sendo implicitamente representada por zeros em todas as novas colunas. Esta técnica é empregada para evitar a multicolinearidade perfeita (também conhecida como "armadilha da variável dummy"), um problema comum em modelos de regressão linear onde a soma das variáveis dummy para uma categoria seria sempre 1, criando uma dependência linear exata.

import pandas as pd
import numpy as np

# Re-utilizar o DataFrame df_categorical e df_one_hot.
# (Recriados aqui para garantir que o bloco seja executável de forma independente se necessário)
if 'df_categorical' not in locals() or df_categorical.empty:
    data = {
        'Sex': ['M', 'F', 'M', 'F', 'M', 'M', 'F'],
        'Equipment': ['Raw', 'Wraps', 'Raw', 'Single-ply', 'Raw', 'Wraps', 'Raw'],
        'Federation': ['IPF', 'USAPL', 'IPF', 'WPC', 'USAPL', 'IPF', 'WPC'],
        'AgeClass': ['24-34', '35-39', '24-34', '40-44', '24-34', '20-23', '35-39']
    }
    df_categorical = pd.DataFrame(data).dropna().copy()
    print("DataFrame de exemplo recriado.")
# Recriar df_one_hot para garantir que esteja disponível para comparação
df_one_hot = pd.get_dummies(df_categorical, prefix=df_categorical.columns)


print("\n--- b) Vetorização Usando Dummy Coding ---")
# Aplicar Dummy Coding (usando drop_first=True para remover a primeira categoria como referência)
df_dummy = pd.get_dummies(df_categorical, prefix=df_categorical.columns, drop_first=True)

print("\nResultados do Dummy Coding (primeiras 5 linhas):")
print(df_dummy.head().to_string())

print(f"\nDimensões do DataFrame após Dummy Coding: {df_dummy.shape}")
print(f"Colunas geradas: {list(df_dummy.columns)}")

print("\n--- Comparação com One-Hot Encoding (a) ---")
print(f"Dimensões One-Hot Encoding: {df_one_hot.shape}")
print(f"Dimensões Dummy Coding: {df_dummy.shape}")

print("\nExemplo de colunas geradas para 'Sex':")
sex_cols_one_hot = [col for col in df_one_hot.columns if col.startswith('Sex_')]
sex_cols_dummy = [col for col in df_dummy.columns if col.startswith('Sex_')]

print(f"One-Hot: {sex_cols_one_hot}")
print(f"Dummy: {sex_cols_dummy}")

print("\n**Observação sobre a comparação:**")
print("O Dummy Coding resulta em K-1 colunas para uma variável com K categorias, reduzindo a dimensionalidade em uma coluna por variável em comparação com One-Hot Encoding. A categoria omitida serve como base para comparação.")



#Vetorize as variáveis categóricas usando Effect Coding. Compare os resultados desta vetorização com aqueles obtidos nos itens (a) e (b).


import pandas as pd
import numpy as np

# Re-utilizar o DataFrame df_categorical, df_one_hot e df_dummy.
# (Recriados aqui para garantir que o bloco seja executável de forma independente se necessário)
if 'df_categorical' not in locals() or df_categorical.empty:
    data = {
        'Sex': ['M', 'F', 'M', 'F', 'M', 'M', 'F'],
        'Equipment': ['Raw', 'Wraps', 'Raw', 'Single-ply', 'Raw', 'Wraps', 'Raw'],
        'Federation': ['IPF', 'USAPL', 'IPF', 'WPC', 'USAPL', 'IPF', 'WPC'],
        'AgeClass': ['24-34', '35-39', '24-34', '40-44', '24-34', '20-23', '35-39']
    }
    df_categorical = pd.DataFrame(data).dropna().copy()
    print("DataFrame de exemplo recriado.")
# Recriar df_one_hot e df_dummy para garantir que estejam disponíveis
df_one_hot = pd.get_dummies(df_categorical, prefix=df_categorical.columns)
df_dummy = pd.get_dummies(df_categorical, prefix=df_categorical.columns, drop_first=True)


print("\n--- c) Vetorização Usando Effect Coding ---")

# Função para aplicar Effect Coding
def apply_effect_coding(df, column):
    # Gera o One-Hot Encoding completo primeiro
    one_hot_df = pd.get_dummies(df[column], prefix=column)

    # Converte para int para permitir valores -1 e evitar FutureWarning
    effect_coded_df = one_hot_df.astype(int)

    # Identifica a última coluna (que será a categoria de referência a ser codificada como -1)
    # Por padrão, get_dummies ordena as colunas alfabeticamente. A última será a referência.
    reference_col_name = effect_coded_df.columns[-1]

    # Para as linhas onde a categoria de referência era True (1), definimos -1 nas outras colunas
    # Iteramos sobre todas as colunas geradas pelo one-hot, exceto a de referência
    for col in effect_coded_df.columns:
        if col != reference_col_name:
            # Atribui -1 nas linhas onde a categoria de referência está presente
            effect_coded_df.loc[one_hot_df[reference_col_name] == True, col] = -1

    # Remove a coluna de referência original
    effect_coded_df = effect_coded_df.drop(columns=[reference_col_name])

    return effect_coded_df

# Aplicar Effect Coding para cada coluna categórica
df_effect = pd.DataFrame(index=df_categorical.index)
for col in df_categorical.columns:
    effect_coded_part = apply_effect_coding(df_categorical, col)
    df_effect = pd.concat([df_effect, effect_coded_part], axis=1)

print("\nResultados do Effect Coding (primeiras 5 linhas):")
print(df_effect.head().to_string())

print(f"\nDimensões do DataFrame após Effect Coding: {df_effect.shape}")
print(f"Colunas geradas: {list(df_effect.columns)}")

print("\n--- Comparação com One-Hot Encoding (a) e Dummy Coding (b) ---")
print(f"Dimensões One-Hot Encoding: {df_one_hot.shape}")
print(f"Dimensões Dummy Coding: {df_dummy.shape}")
print(f"Dimensões Effect Coding: {df_effect.shape}")

print("\nExemplo de colunas geradas para 'Sex':")
sex_cols_one_hot = [col for col in df_one_hot.columns if col.startswith('Sex_')]
sex_cols_dummy = [col for col in df_dummy.columns if col.startswith('Sex_')]
sex_cols_effect = [col for col in df_effect.columns if col.startswith('Sex_')]

print(f"One-Hot: {sex_cols_one_hot}")
print(f"Dummy: {sex_cols_dummy}")
print(f"Effect: {sex_cols_effect}")

print("\nValores para a variável 'Sex' (comparativo):")
print("\nOne-Hot Encoding:")
print(df_one_hot[[col for col in df_one_hot.columns if col.startswith('Sex_')]].head().to_string())
print("\nDummy Coding (considerando a primeira categoria como referência):") # Ex: 'F' como referência se 'Sex_M' for a coluna
print(df_dummy[[col for col in df_dummy.columns if col.startswith('Sex_')]].head().to_string())
print("\nEffect Coding (considerando a última categoria como referência):") # Ex: 'M' como referência se 'Sex_F' for a coluna gerada
print(df_effect[[col for col in df_effect.columns if col.startswith('Sex_')]].head().to_string())


print("\n--- Observações Finais sobre as Comparações ---")
print("1. **Dimensionalidade:** Tanto o **Dummy Coding** quanto o **Effect Coding** resultam em K-1 colunas para K categorias, reduzindo a dimensionalidade em comparação com o **One-Hot Encoding** completo.")
print("2. **Multicolinearidade:** Ambos **Dummy Coding** e **Effect Coding** são projetados para evitar a multicolinearidade perfeita que pode surgir com o One-Hot Encoding, tornando-os mais adequados para modelos que são sensíveis a essa propriedade (ex: regressão linear).")
print("3. **Interpretação dos Coeficientes:**")
print("   - **One-Hot Encoding:** Cada coeficiente representa o efeito daquela categoria específica em relação a uma linha de base implícita (intercepto).")
print("   - **Dummy Coding:** Os coeficientes representam a diferença em relação à categoria de referência explicitamente omitida.")
print("   - **Effect Coding:** Os coeficientes representam o desvio em relação à média geral (ou média não ponderada) das categorias. Isso pode ser mais intuitivo em certas análises, pois a soma dos 'efeitos' das categorias é zero, e o efeito da categoria de referência é o negativo da soma dos efeitos das categorias explicitamente codificadas.")
print("\nA escolha da técnica de vetorização depende do modelo de Machine Learning a ser utilizado e da necessidade de interpretabilidade dos resultados para o problema em questão.")
```

```python
#O que é uma Projeção Linear?
#Uma projeção linear é reduzir os dados de forma estratégica em um espaço de alta dimensão para um espaço de dimensão menor, mas de forma que os pontos permaneçam em uma linha, plano ou hiperplano. Imagine que você está projetando a sombra de um objeto 3D em uma parede (um plano 2D); essa sombra é uma projeção linear. Em Machine Learning, isso significa transformar seus dados de um conjunto maior de características para um conjunto menor de características, mantendo a relação linear entre elas.


#Como uma Projeção Linear é Aplicada a um Espaço de Dados?
#Quando aplicamos uma projeção linear a um espaço de dados, estamos criando uma nova base (um novo conjunto de eixos) para esses dados, onde os novos eixos são combinações lineares dos eixos originais. Os dados são então representados nessas novas coordenadas de dimensão reduzida. Por exemplo, se você tem dados em 3D, uma projeção linear pode mapeá-los para um plano 2D, onde cada ponto 3D agora tem uma coordenada 2D. O objetivo é geralmente encontrar a projeção que capture a maior parte da variância ou informação dos dados originais.

```

```python
#Considere o Lung Cancer Prediction Dataset



import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import kagglehub
from kagglehub import KaggleDatasetAdapter

# --- Carregamento do Dataset ---
# Certifique-se de instalar: pip install kagglehub[pandas-datasets]
file_path = "survey lung cancer.csv" # Nome do arquivo CSV dentro do dataset do Kaggle

try:
    df = kagglehub.load_dataset(
        KaggleDatasetAdapter.PANDAS,
        "shantanugarg274/lung-cancer-prediction-dataset",
        file_path,
    )
    print("Dataset 'Lung Cancer Prediction Dataset' carregado com sucesso.")
except Exception as e:
    print(f"Erro ao carregar o dataset: {e}")
    print("Verifique se 'kagglehub' está instalado e se o arquivo 'survey lung cancer.csv' existe no dataset.")
    print("Criando um DataFrame de exemplo para demonstração.")
    # DataFrame de exemplo para continuar a execução em caso de falha no carregamento
    data = {
        'GENDER': ['M', 'F', 'M', 'F', 'M', 'F', 'M'],
        'AGE': [60, 45, 70, 55, 62, 48, 75],
        'SMOKING': ['YES', 'NO', 'YES', 'NO', 'YES', 'NO', 'YES'],
        'YELLOW_FINGERS': ['YES', 'NO', 'YES', 'NO', 'YES', 'NO', 'YES'],
        'ANXIETY': ['NO', 'YES', 'NO', 'YES', 'NO', 'YES', 'NO'],
        'PEER_PRESSURE': ['YES', 'NO', 'YES', 'NO', 'YES', 'NO', 'YES'],
        'CHRONIC DISEASE': ['NO', 'YES', 'NO', 'YES', 'NO', 'YES', 'NO'],
        'FATIGUE ': ['YES', 'NO', 'YES', 'NO', 'YES', 'NO', 'YES'],
        'ALLERGY ': ['NO', 'YES', 'NO', 'YES', 'NO', 'YES', 'NO'],
        'WHEEZING': ['YES', 'NO', 'YES', 'NO', 'YES', 'NO', 'YES'],
        'ALCOHOL CONSUMING': ['NO', 'YES', 'NO', 'YES', 'NO', 'YES', 'NO'],
        'COUGHING': ['YES', 'NO', 'YES', 'NO', 'YES', 'NO', 'YES'],
        'SHORTNESS OF BREATH': ['NO', 'YES', 'NO', 'YES', 'NO', 'YES', 'NO'],
        'SWALLOWING DIFFICULTY': ['YES', 'NO', 'YES', 'NO', 'YES', 'NO', 'YES'],
        'CHEST PAIN': ['NO', 'YES', 'NO', 'YES', 'NO', 'YES', 'NO'],
        'LUNG_CANCER': ['YES', 'NO', 'YES', 'NO', 'YES', 'NO', 'YES']
    }
    df = pd.DataFrame(data)

print("\nPrimeiras 5 linhas do dataset original:")
print(df.head())
print("\nInformações iniciais do dataset:")
df.info()

# --- Pré-processamento de Dados para PCA ---
# Separar features (X) e target (y)
X = df.drop('LUNG_CANCER', axis=1).copy()
y = df['LUNG_CANCER'].copy()

# Limpar nomes das colunas (remover espaços em branco no início/fim)
X.columns = X.columns.str.strip()

# Mapear variáveis categóricas binárias para 0 e 1
X['GENDER'] = X['GENDER'].map({'M': 1, 'F': 0})

# Colunas que são YES/NO (todas exceto GENDER e AGE, e LUNG_CANCER que é o target)
binary_cols_to_map = [col for col in X.columns if X[col].dtype == 'object']
for col in binary_cols_to_map:
    X[col] = X[col].map({'YES': 1, 'NO': 0})

# Mapear a variável alvo para 0 e 1
y = y.map({'YES': 1, 'NO': 0})

# Padronizar a feature numérica 'AGE'
scaler = StandardScaler()
X['AGE'] = scaler.fit_transform(X[['AGE']])

print("\nPrimeiras 5 linhas do DataFrame após pré-processamento:")
print(X.head())
print("\nVerificando tipos de dados após pré-processamento:")
X.info()


##################################

# --- Aplicação do PCA ---
pca = PCA() # Inicializa PCA para capturar todas as componentes
pca.fit(X)

# Variância explicada por cada componente
explained_variance_ratio = pca.explained_variance_ratio_
# Variância explicada acumulada
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

print("\n--- Análise de Componentes Principais ---")
print("Variância Explicada por Componente Individual:")
for i, var in enumerate(explained_variance_ratio):
    print(f"PC {i+1}: {var:.4f}")

print("\nVariância Explicada Acumulada:")
for i, cum_var in enumerate(cumulative_explained_variance):
    print(f"PC {i+1}: {cum_var:.4f}")

# Determinar o número de componentes para explicar 95% da variância
n_components_95_var = np.where(cumulative_explained_variance >= 0.95)[0][0] + 1
print(f"\nNúmero de Componentes Principais necessárias para explicar >= 95% da variância: {n_components_95_var}")

# Reaplicar PCA com o número selecionado de componentes
pca_final = PCA(n_components=n_components_95_var)
X_pca = pca_final.fit_transform(X)

print(f"\nDimensão dos dados após PCA com {n_components_95_var} componentes: {X_pca.shape}")


# --- Visualização Gráfica dos Resultados ---

# Gráfico 1: Scree Plot (Variância Explicada)
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', linestyle='--', color='blue', label='Variância Explicada Individual')
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='-', color='red', label='Variância Explicada Acumulada')
plt.axvline(x=n_components_95_var, color='green', linestyle=':', label=f'{n_components_95_var} PCs (>= 95% Variância)')
plt.title('Gráfico de Variância Explicada por Componentes Principais (Scree Plot)')
plt.xlabel('Número da Componente Principal')
plt.ylabel('Variância Explicada / Acumulada')
plt.grid(True)
plt.legend()
plt.xticks(range(1, len(explained_variance_ratio) + 1)) # Garante ticks para cada PC
plt.tight_layout()
plt.show()

# Gráfico 2: Projeção 2D das Componentes Principais (PC1 vs PC2)
if X_pca.shape[1] >= 2:
    plt.figure(figsize=(10, 7))
    sns.scatterplot(
        x=X_pca[:, 0], # Primeira Componente Principal
        y=X_pca[:, 1], # Segunda Componente Principal
        hue=y,         # Colorir por LUNG_CANCER (0=Não, 1=Sim)
        palette='viridis', # Paleta de cores
        alpha=0.7,
        s=50
    )
    plt.title(f'Projeção 2D das Componentes Principais (PC1 vs PC2) - {n_components_95_var} PCs Selecionadas')
    plt.xlabel('Componente Principal 1')
    plt.ylabel('Componente Principal 2')
    plt.grid(True)
    plt.legend(title='Câncer de Pulmão', labels=['Não', 'Sim'])
    plt.tight_layout()
    plt.show()
elif X_pca.shape[1] == 1:
    print("\nApenas 1 componente principal foi selecionada. Não é possível criar um scatter plot 2D.")
    plt.figure(figsize=(8, 4))
    sns.histplot(X_pca[:, 0], kde=True, bins=20, color='skyblue')
    plt.title('Distribuição da Componente Principal 1')
    plt.xlabel('Componente Principal 1')
    plt.ylabel('Frequência')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("\nNão há componentes suficientes para visualização 2D ou 1D.")
```
